## Regularization <br/>

This is used to overcome Overfitting and helps to get generalised Hypothesis function.<br/>
Overfitting occurs when we have too many features . In this case Hypothesis may fit the training set very well <br/>
But fails to generalize to new example.

There are two ways to address Overfitting:

1. reduce the number of features either manually or model selection algorithm 

2. Regularization : Keep all the features but reduce the magnitude of Theta. <br/>
Regularization works well when we have lots of slightly useful data or features having less significant.


```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = os.getcwd() + '\..\data\ex2data2.txt'
mydata = pd.read_csv(path, header = None , names = ['Test 1','Test 2','Accepted'])

positive = mydata[mydata['Accepted'].isin([1])]
negative = mydata[mydata['Accepted'].isin([0])]

fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(positive['Test 1'],positive['Test 2'],s =50, c= 'b', marker ='o', label='Accepted')
ax.scatter(negative['Test 1'], negative['Test 2'], s= 50 , c= 'r' , marker ='x', label = 'Rejected')
ax.legend()
ax.set_xlabel('Test 1 Score')
ax.set_ylabel('Test 2 Score')
```

```python
def sigmoid(z):
    return 1 / (1+np.exp(-z))
```

In the above case we can see that there is no linear decision bowndary , <br/>
so what we can do here is construct features derived from Polynomial of the original features.

```python
degree = 5  
x1 = mydata['Test 1']  
x2 = mydata['Test 2']

mydata.insert(3, 'Ones', 1)

for i in range(1, degree):  
    for j in range(0, i):
        mydata['F' + str(i) + str(j)] = np.power(x1, i-j) * np.power(x2, j)

mydata.drop('Test 1', axis=1, inplace=True)  
mydata.drop('Test 2', axis=1, inplace=True)

mydata.head()  
```

```python
def get_regularized_cost(theta, X, y,  reagularization_parameter):
    X = np.matrix(X)
    y = np.matrix(y)
    theta = np.matrix(theta)
    #when y = 1
    first = np.multiply(-y , np.log(sigmoid(X * theta.T)))
    
    #when y = 0
    second = np.multiply((1 - y) , np.log(1 - sigmoid(X * theta.T)))
    
    #note that we are not regularizing the theta zero here.
    reg_term = reagularization_parameter / (2*len(X)) * np.sum(np.power(theta[:,1:theta.shape[1]],2))
    regularized_cost = (np.sum(first - second) / len(X)) + reg_term
    return regularized_cost
```

Notice we have added reagularization_parameter in the above function.If reagularization_parameter is too big then value of theta will be too small , then we might face Underfitting . Similarly if reagularization_parameter is too small then function might face overfittting. 
Now we will add the regularization to the Gradient function.<br/>

```python
def get_regularized_gradient( theta, X, y, learningRate):
    X = np.matrix(X)
    y = np.matrix(y)
    theta = np.matrix(theta)
    parameters = int(theta.ravel().shape[1])
    grad = np.zeros(parameters)
    error = sigmoid(X * theta.T) - y
   
    for i in range(parameters):
        term = np.multiply(error, X[:,i])
        if (i == 0):
            grad[i] = np.sum(term) / len(X)
        else:
            #because we are not regularizing the theta zero
            grad[i] = (np.sum(term) / len(X)) + ((learningRate / len(X)) * theta[:,i])

    return grad                      
```

```python
cols  = mydata.shape[1]
X = mydata.iloc[:,1:cols]
y = mydata.iloc[:,0:1]

X = np.array(X.values)
y = np.array(y.values)
theta = np.zeros(11)

#Regularization saves us from Overfitting. If reagularization_parameter is much bigge than function will be underfitted.
reagularization_parameter = 1

print(get_regularized_cost(theta, X, y,  reagularization_parameter))
print(get_regularized_gradient( theta, X, y, reagularization_parameter))

```

```python
import scipy.optimize as opt
result2 = opt.fmin_tnc(func=get_regularized_cost, x0=theta, fprime=get_regularized_gradient, args=(X, y,reagularization_parameter))
result2
#from scipy.optimize import check_grad
#print(check_grad(get_regularized_cost, get_regularized_gradient, np.random.random(size=20), X, y,0.001))
```

```python
theta_min = np.matrix(result2[0]) 
get_regularized_cost(theta_min, X, y,  reagularization_parameter)
```

```python
def predict(theta, X): 
    
    probability = sigmoid(X * theta.T)
    
    return [1 if x >= 0.5 else 0 for x in probability]
```

```python
predictions = predict(theta_min, X)  
correct = [1 if ((a == 1 and b == 1) or (a == 0 and b == 0)) else 0 for (a, b) in zip(predictions, y)]  
accuracy =  (sum(correct)*100)/len(correct)
print(accuracy)
```
