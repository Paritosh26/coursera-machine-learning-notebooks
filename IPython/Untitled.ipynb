## Regularization <br/>

This is used to overcome Overfitting and helps to get generalised Hypothesis function.<br/>
Overfitting occurs when we have too many features . In this case Hypothesis may fit the training set very well <br/>
But fails to generalize to new example.

There are two ways to address Overfitting:

1. reduce the number of features either manually or model selection algorithm 

2. Regularization : Keep all the features but reduce the magnitude of Theta. <br/>
Regularization works well when we have lots of slightly useful data or features having less significant.


```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = os.getcwd() + '\..\data\ex2data2.txt'
mydata = pd.read_csv(path, header = None , names = ['Test 1','Test 2','Accepted'])

positive = mydata[mydata['Accepted'].isin([1])]
negative = mydata[mydata['Accepted'].isin([0])]

fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(positive['Test 1'],positive['Test 2'],s =50, c= 'b', marker ='o', label='Accepted')
ax.scatter(negative['Test 1'], negative['Test 2'], s= 50 , c= 'r' , marker ='x', label = 'Rejected')
ax.legend()
ax.set_xlabel('Test 1 Score')
ax.set_ylabel('Test 2 Score')
```

In the above case we can see that there is no linear decision bowndary , <br/>
so what we can do here is construct features derived from Polynomial of the original features.
